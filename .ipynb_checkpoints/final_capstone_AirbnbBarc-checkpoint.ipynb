# Understanding & Cleaning Data

```python
#looking to find out first what columns have null values
#using 'sum' function will show us how many nulls are found in each column in dataset
listings.isnull().sum()
```

Let's proceed with removing columns 
that are not important and handling of missing data.

```python
listings.drop(['id','host_name','last_review'], axis=1, inplace=True)
```

```python
##Display the new dataframe
listings.info()
```

```python
#replacing all NaN values in 'reviews_per_month' with 0
listings.fillna({'reviews_per_month':0}, inplace=True)
#examing changes
listings.reviews_per_month.isnull().sum()
```

```python
#let's find out more about our neiberhoods presented 

#Eixample
sub_1=listings.loc[listings['neighbourhood_group'] == 'Eixample']
price_sub1=sub_1[['price']]
#Sant Martí
sub_2=listings.loc[listings['neighbourhood_group'] == 'Sant Martí']
price_sub2=sub_2[['price']]
#Gràcia
sub_3=listings.loc[listings['neighbourhood_group'] == 'Gràcia']
price_sub3=sub_3[['price']]
#Les Corts
sub_4=listings.loc[listings['neighbourhood_group'] == 'Les Corts']
price_sub4=sub_4[['price']]
#Ciutat Vella
sub_5=listings.loc[listings['neighbourhood_group'] == 'Ciutat Vella']
price_sub5=sub_5[['price']]
#Sants-Montjuïc
sub_6=listings.loc[listings['neighbourhood_group'] == 'Sants-Montjuïc']
price_sub6=sub_6[['price']]
#Sarrià-Sant Gervasi
sub_7=listings.loc[listings['neighbourhood_group'] == 'Sarrià-Sant Gervasi']
price_sub7=sub_7[['price']]
##Horta-Guinardó
sub_8=listings.loc[listings['neighbourhood_group'] == 'Horta-Guinardó']
price_sub8=sub_8[['price']]
#Sant Andreu
sub_9=listings.loc[listings['neighbourhood_group'] == 'Sant Andreu']
price_sub9=sub_9[['price']]
#Nou Barris
sub_10=listings.loc[listings['neighbourhood_group'] == 'Nou Barris']
price_sub10=sub_10[['price']]
#putting all the prices' dfs in the listings
p_l_b_n=[price_sub1, price_sub2, price_sub3, price_sub4, price_sub5, price_sub6, price_sub7, price_sub8, price_sub9, price_sub10]
```

```python
s1 = price_sub1.describe(percentiles=[.25,.5,.75])
s1.info()
```

```python
#creating an empty list that we will append later with price distributions for each neighbourhood_group
p_l_b_n_2=[]
#creating list with known values in neighbourhood_group column
nei_list=['Eixample', 'Sant Martí', 'Gràcia', 'Les Corts', 'Ciutat Vella', 'Sants-Montjuïc', 'Sarrià-Sant Gervasi', 'Horta-Guinardó', 'Sant Andreu', 'Nou Barris']
#creating a for loop to get statistics for price ranges and append it to our empty list
for x in p_l_b_n:
    i=x.describe(percentiles=[.25, .50, .75])
    i=i.iloc[3:]
    i.reset_index(inplace=True)
    i.rename(columns={'index':'Stats'}, inplace=True)
    p_l_b_n_2.append(i)
#changing names of the price column to the area name for easier reading of the table    
p_l_b_n_2[0].rename(columns={'price':nei_list[0]}, inplace=True)
p_l_b_n_2[1].rename(columns={'price':nei_list[1]}, inplace=True)
p_l_b_n_2[2].rename(columns={'price':nei_list[2]}, inplace=True)
p_l_b_n_2[3].rename(columns={'price':nei_list[3]}, inplace=True)
p_l_b_n_2[4].rename(columns={'price':nei_list[4]}, inplace=True)
p_l_b_n_2[5].rename(columns={'price':nei_list[5]}, inplace=True)
p_l_b_n_2[6].rename(columns={'price':nei_list[6]}, inplace=True)
p_l_b_n_2[7].rename(columns={'price':nei_list[7]}, inplace=True)
p_l_b_n_2[8].rename(columns={'price':nei_list[8]}, inplace=True)
p_l_b_n_2[9].rename(columns={'price':nei_list[9]}, inplace=True)


#finilizing our dataframe for final view    
stat_df=p_l_b_n_2
stat_df=[df.set_index('Stats') for df in stat_df]
stat_df=stat_df[0].join(stat_df[1:])
stat_df
```
